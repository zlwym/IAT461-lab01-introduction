### Part 7: Mapping and Transformation ###

```python
#2
categories = df['Age'].map(
    lambda x: 'Child' if pd.notnull(x) and x < 18
    else ('Adult' if pd.notnull(x) and x < 60
          else 'Senior' if pd.notnull(x) else np.nan))

df['Age Group'] = categories
display(df)
```

**Explanation**

Also conditionally searches for null values and returns rows with null ages to null age groups (using if pd.notnull(x) and else np.nan). Creates a new column called 'Age Group' and assigns it to these new categories.

### Part 8: Part 3 in Plain Python ###
1. Show only the rows where the age is above 60.


```python
import csv
import urllib.request

with urllib.request.urlopen("https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/titanic.csv") as response:
    lines = [line.decode("utf-8") for line in response.readlines()]
    reader = csv.DictReader(lines)
    dataList = list(reader)

overSixty = []

for row in dataList:
    age = row["Age"]

    if age != "":
        age = float(age)
        if age > 60:
            overSixty.append(row)

for p in overSixty:
    print(p)
```

**Explanation**

I used csv and which is a built-in Python module and urllib to read the url. These are both part of Python's standard library and are not external. I first used urllib to open the url and csv to read the csv. Then, I converted the csv into a list of dictionaries.

I created an empty list for the rows of passengers above 60. Using a for loop I visit each row's 'Age' column, check if it is empty, convert it to a float, and then add the row to the empty list if the age is over 60. Then I print each row.

### Part 8: Part 5 in Plain Python ###
1. Create a new column that converts fares from USD to Euros.


```python
import csv
import urllib.request

with urllib.request.urlopen("https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/titanic.csv") as response:
    lines = [line.decode("utf-8") for line in response.readlines()]
    reader = csv.DictReader(lines)
    dataList = list(reader)

for row in dataList:
    fare = row["Fare"]

    if fare != "":
        fare_usd = float(fare)
        fare_eur = fare_usd * 0.86
        row["Fare (Euros)"] = round(fare_eur, 2)
    else:
        row["Fare (Euros)"] = ""

for row in dataList:
    print(row)
```

**Explanation**

Used csv and urllib to open and read the url and csv, turning it into a list of dictionaries. For each row, the 'Fare' column is selected. If it is not null, the fare will be multiplied by the exchange rate and rounded to 2 decimal places. It gets added to "Fare (Euros)". If the value is null, a null value gets returned for Euros.
