# Basic Text Styling #

### Bold ###

\*\*Bold Text\*\*  \_\_Bold Text\_\_

Do not add a space after the * or _ sign forbold text.

example: **This text is Bold**

### Italic ###

\*Italic text\* \_Italic text\_

No space after the sign.

example: *this text is italic.*

### Bold and Italic ###

\*\*\*Bold and Italic text\*\*\*

example: ***this text is bold and italic.***

### Strikethrough ###

\~\~strikethrough text\~\~

example: ~~this is a strikethrough text~~

### Headings ###

For heading, add a sapce after the sign.

\# Large Heading \#

\#\# Medium Heading \#\#

\#\#\# Small Heading \#\#\#

example:
# H1 #

## H2 ##

### H3 ###

### Lists ###

### Unordered ###
- Item1
- Item2
    - SubItem

### Ordered ###
1. first
2. second
    1. sub-item
