# Pandas #

Pandas is a powerful Python library for handling and analyzing data in table format. It makes it easy to manipulate structured data—especially for data science tasks.

## Create a Sample DataFrame ## 

```python
# First importing the pandas and numpy libraires
import pandas as pd
import numpy as np
```

```python
# Creating two list for car makes and months

# a numpy array
products = np.array(['Corolla', 'Camry', 'RAV4', 'Highlander', 'Prius', 'Tacoma'])

# a normal array
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
# Sample monthly sales data (in $1000s)
sales_data = np.array([
    [23, 25, 22, 20, 30, 28, 26, 27, 29, 33, 31, 35],
    [15, 18, 16, 20, 22, 21, 19, 17, 18, 22, 24, 26],
    [35, 33, 30, 28, 36, 34, 32, 38, 37, 39, 40, 42],
    [12, 15, 14, 13, 16, 18, 17, 14, 15, 16, 18, 19],
    [28, 30, 32, 29, 31, 33, 35, 34, 36, 38, 37, 40],
    [20, 22, 21, 19, 23, 25, 24, 22, 23, 25, 26, 27]
])
```

```python
# Create the DataFrame and print the first five data rows
df = pd.DataFrame(sales_data, index=products, columns=months)
print(df.head())
```

### Selecting a Column ###
You can select a single column using bracket notation ([]):

**Note**: When you select a single column like this, Pandas returns a **Series**.
A Series is like a single column of a table. It's a **one-dimensional labeled array** that holds data and an associated index.

```python
# Sales in January (returns a Series)
df['Jan']
```

- The values are the January sales numbers.

- The index is the list of car models.

- The name of the Series is "Jan".

If you want to keep your result in dataFrame format, you can select it like this instead:

```python
df[['Jan']]
```

This returns a DataFrame with just one column instead of a Series. Or you can call the **to_frame()** method.

```python
df['Jan'].to_frame()
```

### Selecting Rows by Label (.loc[]) ###
The .loc[] method is used to select rows (and optionally columns) by their labels:

```python
# Get all monthly sales for Camry (returns a series)
df.loc['Camry']
```

### Selecting a Specific Row and Column ###
You can use df.loc[row_label, column_label] to access a single value in the DataFrame. This is similar to accessing a specific cell in a spreadsheet.

```python
# Get Camry's sales in March
df.loc['Camry', 'Mar']
```

Note:The output is not a Series or DataFrame—it’s just a single value (scalar), such as int or float.

### Selecting Multiple Rows and Columns ###
If you want to return a small sub-table (i.e., a DataFrame), you can use a list of labels:

```python
# Sales of Camry and Prius in Jan and Feb
df.loc[['Camry', 'Prius'], ['Jan', 'Feb']]
```

This returns a DataFrame, not a Series

### Selecting Rows by Position (.iloc[]) ###
The .iloc[] method is used to select rows/columns by index position (like in NumPy):

```python
# First row (Corolla)
df.iloc[0].to_frame()
```

```python
# Third row, fifth column (RAV4’s May sales)
df.iloc[2, 4]
```

### Selecting a Range or Subsection (Slicing) ###
Slicing in Pandas allows you to extract a portion of the DataFrame—just like slicing lists or arrays in Python—but with labeled data.

You can select a range of rows using .loc[], where you specify the **start and end row labels** by using **":"**.

```python
# Sales from Corolla to RAV4 (inclusive)
df.loc['Corolla':'RAV4']
```

```python
# Sales in Jan to Mar for all models
df.loc[:, 'Jan':'Mar']
```

### Boolean Indexing (Conditional Selection) ###
Boolean indexing allows you to filter rows (or columns) of a DataFrame based on a condition. It returns only **the rows for which the condition is True**.

```python
#df['Jan'] > 25 returns a Boolean Series (True/False for each row).
#df[...] uses that Boolean Series to return only rows where the condition is True.
# the resulting dataframe is assigned to "tdf"
tdf = df[df['Jan'] > 25]
print(tdf)
```

### Chaining Multiple Conditions ###
- Use & for AND

- Use | for OR

- Each condition must be wrapped in parentheses

- Do not use and/or (those are for single booleans, not Series)

```python
# Models with January > 25 AND February > 30
tdf = df[(df['Jan']>25) & (df['Feb'] > 30)]
print(tdf)
```

```python
#  Models where March < 20 OR December > 40
tdf = df[(df['Mar']<20) | (df['Dec']>40)]
print(tdf)
```

### Reindexing ###
Reindexing means changing the order of existing row or column labels or introducing new ones. This is especially useful when:

- You want to sort or reorder your data.

- You need to align your data to another structure.

- You want to introduce missing entries (e.g., add new car models or months).

**How to use:**
- Use the reindex() method to perform reindexing.

- For rows: df.reindex(new_row_labels)

- For columns: df.reindex(columns=new_col_labels)

- For both: df.reindex(index=..., columns=...)

```python
# let's reorder the rows
new_order = ['RAV4', 'Tacoma', 'Camry', 'Corolla', 'Prius', 'Highlander']
df_reordered = df.reindex(new_order)
print(df_reordered)
```

**What happens:**

Rows are reordered according to new_order.

Data is preserved for existing labels.

If any label is missing in the original, Pandas will insert NaN.

```python
# You can reindex with additional labels to introduce missing rows:
df_expanded = df.reindex(['RAV4', 'Tacoma', 'Yaris', 'Tundra'])
print(df_expanded)
```

```python
# reorder columns
month_order = ['Mar', 'Jan', 'Dec', 'Jun']
df_cols_reordered = df.reindex(columns=month_order)
print(df_cols_reordered)
```

```python
#Reindex with Both Rows and Columns
df_custom = df.reindex(index=['Prius', 'Camry', 'Yaris'], columns=['Feb', 'Aug', 'Oct'])
print(df_custom)
```

### Dropping Entries ###

Dropping means removing rows or columns from a DataFrame. You can drop:

- One or more rows by label (e.g., a car model),

- One or more columns by label (e.g., a month),

- Temporarily (without modifying the original DataFrame), or permanently (by using inplace=True).

**How it works**

<pre>df.drop(labels, axis, inplace=True/False)</pre>

- labels: the names of rows or columns to drop

- axis=0: drop by row

- axis=1: drop by column

- inplace=True: apply changes directly to df (no need to assign, inplace=False (default): retruns a new df

```python
# Drop a Single Row (e.g Camry)
df.drop('Camry') # axis=0 is default
```

```python
# Drop Multiple Rows
df.drop(['Camry', 'Tacoma'])
```

```python
#  Drop a Column (e.g Jan)
df.drop('Jan', axis=1)
```

```python
# drop multiple columns
df.drop(['Jan', 'Feb'], axis=1)
```

**Drop with inplace=True (Destructive Operation use with caution)**

This modifies the original df and returns None.

Always use this cautiously or copy your data first.

```python
df.drop('Camry', inplace=True)
print(df)
```

### Arithmetic and Data Alignment ###

Pandas lets you perform arithmetic (addition, subtraction, etc.) between entire DataFrames or Series, while automatically matching rows and columns by their labels. This is called label-based alignment.

Even if DataFrames or Series are not the same shape, Pandas will align them based on their index and column names and fill in missing values with NaN.

**Basic Arithmetic with Scalars**

```python
# Add 10 to every sales value
df_plus_10 = df + 10
print(df_plus_10)
```

```python
# Multiply all sales by 1.1 (e.g., apply a 10% price increase)
df_bonus = df * 1.1
print(df_bonus)
```

```python
# Monthly average sales which is a series
monthly_avg = df.mean()

# Subtract average from each month's data (column-wise)
df_diff = df - monthly_avg
print(df_diff)
```

**Use .add(), .sub(), .mul(), .div() with fill_value**

To prevent NaNs when labels are missing, use arithmetic methods with fill_value:

```python
df1 = pd.DataFrame({'A': [1, 2]}, index=['X', 'Y'])
df2 = pd.DataFrame({'A': [3]}, index=['X'])

# Default behavior: Y becomes NaN
df1 + df2

# Safe addition with fill value 
df1.add(df2, fill_value=0) #try it without fill_value
```

### Applying Functions ###
Pandas gives you multiple ways to apply functions to your data, depending on whether you want to operate on:

- Whole rows or columns

- Each individual element

The .apply() method is used when you want to apply a function across each row or column.

The .apply() method expects to receive a function as its argument.

A function that takes either a Series (if used on a DataFrame) or a scalar (if used on a Series) and returns a value.

The function can be a
**built-in function**,
**user-defined function**,
or a **lambda function**



```python
# built-in
print(df.apply(np.mean))        # applies to each column by default
```

```python
print(df.apply(np.sum, axis=1)) # applies to each row
```

```python
# user-defined function
def range_func(x):
    return x.max() - x.min()

df_range = df.apply(range_func, axis=1)  # applies to each row
print(df_range)
```

```python
# lambda
df.apply(lambda c: c.mean()) # column wise
```

###  Mapping ###
.map() is used for element-wise transformations on Series. It's most useful when:

- You want to transform values based on rules

- You want to replace values based on a mapping

```python
# Categorize Car Models by Performance

# Average sales per model
avg_sales = df.mean(axis=1)

# Map numeric average to label
categories = avg_sales.map(lambda x: 'High' if x > 30 else 'Low')
print(categories)
```
