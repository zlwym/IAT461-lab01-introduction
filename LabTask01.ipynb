### Part 3: Filtering ###

1. Show only the rows where the age is above 60.

2. Show rows where the fare is above 100 and the passenger was in 1st class.

3. Identify and isolate rows with missing values in the age column.

```python
#1
ageOverSixty = df[df['Age'] > 60]
display(ageOverSixty)
```

```python
#2
fareOverOneHundred = df[(df['Fare'] > 100) & (df['Pclass'] == 1)]
display(fareOverOneHundred)
```

```python
#3
nullAge = df[df['Age'].isnull()]
display(nullAge)
```

### Part 4: Reindexing and Dropping ###

```python
#1
reversed = df.reindex(index=df.index[::-1])
display(reversed)
```

```python
#3
noNullAge = df.dropna(subset = ["Age"])
display(noNullAge)
```
